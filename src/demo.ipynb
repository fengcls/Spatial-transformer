```python
%matplotlib inline 
import numpy as np
import matplotlib.pyplot as plt

fcn_ac = [0.6056,0.8648,0.9821,0.941,0.8093]
stfcn_ac = [0.8934,0.9658,0.9877,0.9628,0.9675]
plt.figure(figsize=(9,9))
plt.plot(1-np.array(fcn_ac), label='Error without transform layer')
plt.plot(1-np.array(stfcn_ac), label='Error with transform layer')
plt.legend(fontsize=18,loc=1)
plt.xlabel('distortion', fontsize=20)
plt.ylabel('error', fontsize=20)
plt.title('Error Rate for FCN',fontsize=25)
plt.xticks(range(5),('TC','T','P','R','RTS'),fontsize=15)
plt.show()
plt.savefig('fcnstfcn.png')
```
